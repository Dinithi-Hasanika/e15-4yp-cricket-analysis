```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#load data file
data = pd.read_csv('allMatchData.csv')
```

```python
#load data file
Batsmen = pd.read_csv('BtScore1.csv')
```

```python
players = ['p1','p2','p3','p4','p5','p6','p7','p8','p9','p10','p11']
```

```python
sumslist = []
```

```python
 for index, row in data.iterrows():
        sums = 0
        for ind, p in enumerate(players):
            player = data.iloc[index][p]
            value = Batsmen[Batsmen['Name']==player]['Batsmen_score2'].values
            
            try:
                sums = sums + value[0]
            except IndexError:
                pass
                
        print(sums)
        sumslist.append(sums)
```

```python
data['Batsmen Score'] = sumslist
```

```python
#load data file
Bowlers = pd.read_csv('BwScore1.csv')
```

```python
sumslist = []
for index, row in data.iterrows():
        sums = 0
        for ind, p in enumerate(players):
            player = data.iloc[index][p]
            value = Bowlers[Bowlers['Name']==player]['Bowlers_score'].values
            
            try:
                sums = sums + value[0]
            except IndexError:
                pass
                
        print(sums)
        sumslist.append(sums)
```

```python
data['Bowlers Score'] = sumslist
```

```python
#load data file
Fielders = pd.read_csv('fielders.csv')
```

```python
sumslist = []
for index, row in data.iterrows():
        sums = 0
        for ind, p in enumerate(players):
            player = data.iloc[index][p]
            value = Fielders[Fielders['Name']==player]['Fielders_score'].values
            
            try:
                sums = sums + value[0]
            except IndexError:
                pass
                
        print(sums)
        sumslist.append(sums)
```

```python
data['Fielders Score'] = sumslist
```

```python
from sklearn.preprocessing import LabelEncoder

label = LabelEncoder()
data['Team'] = label.fit_transform(data['Team'])
data['Day_Night'] = label.fit_transform(data['Day_Night'])
data['Ground'] = label.fit_transform(data['Ground'])
data['Toss'] = label.fit_transform(data['Toss'])
data['Bat'] = label.fit_transform(data['Bat'])
data['Opposition'] = label.fit_transform(data['Opposition'])
```

```python
feature_names = ['Team', 'Day_Night', 'Home', 'Ground', 'Toss', 'Bat', 'Opposition', 'Batsmen Score', 'Bowlers Score', 'Fielders Score']
X = data[feature_names]
Y = data['Result']
```

```python
#split the data set as training set and test set randomly
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, random_state=0)
```

```python
#apply scaling
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
#use model Logistic Regression
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression()
logreg.fit(X_train, y_train)

print('Accuracy of Logistic regression classifier on training set: {:.2f}'
     .format(logreg.score(X_train, y_train)))
print('Accuracy of Logistic regression classifier on test set: {:.2f}'
     .format(logreg.score(X_test, y_test)))
# get importance
importance = logreg.coef_[0]
# summarize feature importance
for i,v in enumerate(importance):
	print('Feature:%0d -> %s, Score: %.5f' % (i,feature_names[i],v))
# plot feature importance
plt.bar([x for x in range(len(importance))], importance)
plt.xticks(np.arange(len(feature_names)), feature_names,rotation='vertical')
plt.show()
```

```python
#use model Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier

clf = DecisionTreeClassifier().fit(X_train, y_train)

print('Accuracy of Decision Tree classifier on training set: {:.2f}'
     .format(clf.score(X_train, y_train)))
print('Accuracy of Decision Tree classifier on test set: {:.2f}'
     .format(clf.score(X_test, y_test)))
```
