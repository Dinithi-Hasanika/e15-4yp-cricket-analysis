```python
batList = []
bowlList = []
fieldList = []
for index, row in data.iterrows():
    inning = data.iloc[index]['Innings ID']
    if(inning%2 == 1):
        inning = inning + 1
    else:
        inning = inning - 1
    batList.append(data[data['Innings ID']==inning]['Batsmen Score'].values[0])
    bowlList.append(data[data['Innings ID']==inning]['Bowlers Score'].values[0])
    fieldList.append(data[data['Innings ID']==inning]['Fielders Score'].values[0])
```

```python
data['Opposite Batsmen Score'] = batList
data['Opposite Bowlers Score'] = bowlList
data['Opposite Fielders Score'] = fieldList
```

```python
from sklearn.preprocessing import LabelEncoder

label = LabelEncoder()
data['Team'] = label.fit_transform(data['Team'])
data['Day_Night'] = label.fit_transform(data['Day_Night'])
data['Ground'] = label.fit_transform(data['Ground'])
data['Toss'] = label.fit_transform(data['Toss'])
data['Bat'] = label.fit_transform(data['Bat'])
data['Opposition'] = label.fit_transform(data['Opposition'])
```

```python
feature_names = ['Team', 'Day_Night', 'Home', 'Ground', 'Toss', 'Bat', 'Opposition', '2-gram-bat', 'Batsmen Score', 'Bowlers Score', 'Fielders Score', 'Opposite Batsmen Score', 'Opposite Bowlers Score', 'Opposite Fielders Score']
X = data[feature_names]
Y = data['Result']
```

```python
#split the data set as training set and test set randomly
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, random_state=None)
```

```python
#apply scaling
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
#use model Logistic Regression
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression()
logreg.fit(X_train, y_train)

print('Accuracy of Logistic regression classifier on training set: {:.2f}'
     .format(logreg.score(X_train, y_train)))
print('Accuracy of Logistic regression classifier on test set: {:.2f}'
     .format(logreg.score(X_test, y_test)))
# get importance
importance = logreg.coef_[0]
# summarize feature importance
for i,v in enumerate(importance):
	print('Feature:%0d -> %s, Score: %.5f' % (i,feature_names[i],v))
# plot feature importance
plt.bar([x for x in range(len(importance))], importance)
plt.xticks(np.arange(len(feature_names)), feature_names,rotation='vertical')
plt.show()
```

```python
#use model Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier

clf = DecisionTreeClassifier().fit(X_train, y_train)

print('Accuracy of Decision Tree classifier on training set: {:.2f}'
     .format(clf.score(X_train, y_train)))
print('Accuracy of Decision Tree classifier on test set: {:.2f}'
     .format(clf.score(X_test, y_test)))
```
